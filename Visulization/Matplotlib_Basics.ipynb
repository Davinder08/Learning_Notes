```python
import matplotlib.pyplot as plt
%matplotlib inline 
#replacer of plt.show()
import numpy as np
```

```python
"""# Here we will see following types of graphs
1. Scatter  --- plt.scatter
2. Line     ---- plt.plot
3. Subplot  ---- plt.subplot
4. Barplot  ---- plt.bar
5. Boxplot  ---- plt.boxplot
6. Pie Chart ----
7. Histogram  ---- plt.hist
plt.xlabel/ylabel/xscale/yscale/title/savefig/show
"""
```

```python
x = np.arange(1,10,1)
y = x*2
```

```python
x = np.arange(1,10,1)
y = x*2
plt.scatter(x,y,c='b')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('test')
plt.savefig('Test.png') # saving the graph
```

```python
y=x*x
plt.figure(figsize=(16,8))
plt.subplot(2,4,1)
plt.plot(x,y)
plt.title('plt.plot(x,y)')
plt.subplot(2,4,2)
plt.plot(x,y,'r--')
plt.title('plt.plot(x,y,r--)')
plt.subplot(2,4,3)
plt.plot(x,y,'b*-')
plt.title('plt.plot(x,y,b*-)')
plt.subplot(2,4,4)
plt.plot(x,y,'go')
plt.title('plt.plot(x,y,go)')
plt.subplot(2,4,5)
plt.plot(x,y,'yo--')
plt.title('plt.plot(x,y,yo--)')
plt.subplot(2,4,6)
plt.plot(x,y,'r',linestyle='dashed',linewidth=2,markersize=12)
plt.subplot(2,4,7)
plt.plot(x,y,'b*')
plt.title('plt.plot(x,y,b*)')
plt.subplot(2,4,8)
plt.scatter(x,y,c='g')
plt.title('plt.scatter(x,y,c=g)')
```

```python
x1 = [4,6,8]
y1 = [15,14,16]
x2 = [3,5,7]
y2 = [14,13,17]
plt.bar(x1,y1,color='b')
plt.bar(x2,y2,color='y')
```

```python
a = np.array([1,11,12,21,22,23,31,32,50])
plt.hist(a,bins=5) # by defaultbins are 10
plt.xscale("linear")
```

```python
data = [np.random.normal(0,std,100) for std in range(1,4)] # this will create 3 lists
plt.boxplot(data,vert=True,patch_artist=True) # vert ---> graph as vertical, patch_artist --> Filling
plt.show()
```

```python
labels = 'A','B','C','D'
sizes = [10,20,30,40]
colors = ['pink','pink','pink','pink']
explode = (0.1,0.1,0.2,0.1) # movement
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.0f%%',shadow=True) # %1.2f%% -- float with 2 decimals
plt.show()
```
